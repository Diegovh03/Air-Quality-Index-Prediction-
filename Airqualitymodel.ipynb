### SOLUTION ESTABLISH FOR THE MISSING VALUES


```python
#complete the data wiht mean value , for this we should convert -200 to NaN (not take into account -200 as valuee)
```

```python
air_quality_data = air_quality_data.replace(to_replace= -200, value =np.NaN)
```

```python
air_quality_data.head()
```

```python
air_quality_data.tail()
```

```python
air_quality_data.isnull().sum()
```

```python
air_quality_data.mean(numeric_only=True)
```

```python
#Replacing the missing values with mean  value
air_quality_data = air_quality_data.fillna(air_quality_data.mean(numeric_only=True))
#.fillna(): This DataFrame method is used to fill missing values (NaN) with specific values.
```

### FORECASTING MODEL APPLICATION


```python
#Change date time format according to the application  YYYY-MM-DD HH:MM:SS for a timestamp
#Converting the data from DD/MM/YYYY to YYYY-MM-DD

data_info = pd.to_datetime(air_quality_data['Date'], format="%d/%m/%Y")
print(data_info)
```

```python
time_info = air_quality_data['Time']
```

```python
#Now converting the time from HH.MM.SS to HH:MM:SS

time_info = time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
#Data frame = Several row and Several columns
#Series = one row and one column, not multidimentional

print(type(data_info))
print(type(time_info))
```

```python
date_time = pd.concat([data_info, time_info], axis=1)
```

```python
#combine together in 1 single column 
date_time['ds'] = date_time['Date'].astype(str)+' '+ date_time['Time'].astype(str)
```
